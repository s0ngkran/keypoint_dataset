```python
frame = copy.copy(frame_ori)
pick = [[140,x*100] for x in range(1,7)]
pick.append([10,600])
for y, x in pick:
    thres = 15
    picker = frame[y, x]
    print('rgb=', picker, 'thres=', thres)

    red, green, blue = picker
    val_red = frame_ori[:,:,0]
    val_green = frame_ori[:,:,1]
    val_blue = frame_ori[:,:,2]
    c1 = (val_red>red-thres) * (val_red<red+thres)
    c2 = (val_green>green-thres) * (val_green<green+thres)
    c3 = (val_blue>blue-thres) * (val_blue<blue+thres)
    c = c1*c2*c3
    frame[c] = 0
plt.imshow(frame)
plt.show()
```
